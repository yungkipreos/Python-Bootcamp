My favorite equation is Kepler's Third Law Equation. I like that it's simple and it's cute. Here is the equation:

$P^2 = a^3$

where $P$ is the period of the planet's orbit and a is the semi major axis of the planet's orbit in AU. Below is an example for the Jupyter (lol), which has a semi major axis of 5.2029 AU (https://www.princeton.edu/~willman/planetary_systems/Sol/Jupiter/).



```python
# For the Earth:
a = 5.2029 #AU
P = (a**3)**0.5

print ("Period of Jupiter = ~", round(P, 2), "years")
```
